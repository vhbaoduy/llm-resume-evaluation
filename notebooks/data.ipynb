```python
df = pl.read_csv(
    "../data/dataset/df_final_score_gemini2p5.csv"
).with_row_index().with_columns(_id=pl.concat_str([pl.lit("id_") ,pl.col("index")])).drop("index")
# df = (
#     df
#     .with_columns(
#         resume_path = pl.concat_str([pl.lit(f"{DATA_PATH}/CV"), pl.col("CV")]),
#         jd_path = pl.concat_str([pl.lit(f"{DATA_PATH}/JD"),pl.col("JD")])
#     )
# )
df.write_csv("../data/collect.csv")
```

```python
def prepare_data_for_sentence_classification(
    resume: str, job_description: str, id: str = None
):
    if id is None:
        id = str(uuid.uuid4())
    return [
        {"id": f"resume_{id}", "value": resume},
        {"id": f"jd_{id}", "value": job_description},
    ]


def prepare_data_for_evaluation(resume: str, job_description: str, id: str = None):
    if id is None:
        id = str(uuid.uuid4())

    return {"data_id": id, "resume": resume, "job_description": job_description}


for row in df.to_dicts():
    print(row)
    resume = open(
        os.path.join(
            DATA_PATH,
            "CV",
            row["CV"] + ".txt",
        ),
        "r",
    ).read()
    jd = open(
        os.path.join(
            DATA_PATH,
            "JD",
            row["JD"] + ".txt",
        ),
        "r",
    ).read()
    data = prepare_data_for_sentence_classification(
        resume=resume, job_description=jd, id=row["_id"]
    )
    break
```
